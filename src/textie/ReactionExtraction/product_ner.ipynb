# 使用微调后的llm来识别product实体

```python
import os
from evaluate import *
from llm_prod import *
base_dir=os.getcwd()

def main():
    train_data,test_data=get_productdata()
    prev_data=chem_prod(train_data,test_data)
    evaluater=Product_f1(test_data,prev_data)
    precision,recall,f1=evaluater.getf1()
    f_dict={"precision":precision,"recall":recall,"f1":f1}
    save_dir=base_dir+"/product_result.json"
    write_dict_to_file(f_dict,save_dir)

main()
```
